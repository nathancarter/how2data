This answer assumes you have imported SymPy as follows.

```python
from sympy import *                   # load all math functions
init_printing( use_latex='mathjax' )  # use pretty math output
```

First, we need a two-variable function that we wish to plot.

```python
var( 'x y' )
formula = sin( x**2 + y**2 )
formula
```

You can use `plot3d`, but you have to import it specifically,
because it is not imported by default with the rest of SymPy.

```python
from sympy.plotting.plot import plot3d
plot3d( formula )
```

Specify the minimum and maximum values for both $x$ and $y$ as follows.
In this example, I keep $-\pi\leq x\leq\pi$ and $-\frac\pi2\leq y\leq\frac\pi2$.

```python
plot3d( formula, (x,-pi,pi), (y,-pi/2,pi/2) )
```
